We'll start by loading the data using the pandas library. 
To do this, we'll need to import geopandas, pandas, and matplotlib.pyplot. Then, we'll read in the file containing the school data using the pd.read_csv() function and assign it to the variable school_data.

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# read in the school data file
school_data = pd.read_csv("enrollmentdata.csv")

# display the first 5 rows of the school data
school_data.head(5)
```

We need to focus on two specific columns in the dataset, which are "Neighborhood" and "Estimate; Enrolled in school". By examining these columns, we can identify the neighborhoods with the largest number of enrolled students, and these will be our top candidates.

## Let's sort the school_data in a descending order based on the enrollment column, so that we can analyze the neighborhoods with the highest amounts of enrollment.
We will store the sorted data in a variable called 'plot' and print the top 10 rows.


```python
plot = school_data.sort_values(["Estimate; Total:"], ascending = False)
plot.head(10)
```


The neighborhoods listed on the left are the ones with the highest enrollment and are our top contenders. To visualize this data, let's create a graph.

```python
# sort the data in descending order by enrollment and select the top 10 neighborhoods
top_neighborhoods = school_data.sort_values(["Estimate; Total:"], ascending=False).head(40)

# create bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_neighborhoods['Neighborhood'], top_neighborhoods['Estimate; Total:'], color='purple')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Neighborhood')
plt.ylabel('Enrollment')
plt.title('Top 40 Neighborhoods by School Enrollment')
plt.show()
```

Ranking the neighborhoods in order from best to worse in order

```python
school_data_ranked = school_data.sort_values(by="Estimate; Total:", ascending=False).reset_index(drop=True)
school_data_ranked["Rank"] = school_data_ranked.index + 1

print(school_data_ranked[["Rank", "Neighborhood", "Estimate; Total:"]])
```

```python
# save the sorted DataFrame to a CSV file
school_data_ranked.to_csv("Enrollment_rankings.csv", index=False)
```

## Let's focus on the top 10 neigborhoods since there are a lot of them and it's hard to read the chart.

```python
# create a new dataframe with top 20 neighborhoods by enrollment
top20_neighborhoods = school_data.sort_values(by="Estimate; Total:", ascending=False).head(10)

# plot bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(top20_neighborhoods["Neighborhood"], top20_neighborhoods["Estimate; Total:"], color="purple")
ax.set_xticklabels(top20_neighborhoods["Neighborhood"], rotation=45, ha="right")
ax.set_xlabel("Neighborhood")
ax.set_ylabel("Enrollment")
ax.set_title("Top 10 Pittsburgh neighborhoods by enrollment")
plt.show()
```

## Based on the graph and ranking, it seems that **Squirrel Hill South** has the highest enrollment among all neighborhoods, making it the clear winner.
